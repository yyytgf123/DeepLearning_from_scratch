<H1>numpy</H1>
파이썬에서 과학 계산을 지원하는 라이브러리
특히 배열 연산과 같은 기능이 강력하여 데이터분석, 머신러닝, 인공지능 등의 분야에서 사용

```python
import numpy as np

# numpy 배열 생성
x = np.array([1.0, 2.0, 3.0])
print(x)
```

```python
A = np.array([[1,2],[3,4]])
print(A)
print()
print(A.shape)
```

```python
B = np.array([[1,2],[3,4]])
print(A+B)
print()
print(A*B)
print()
print(A*10)
```

```python
c = np.array([[1,2],[3,4]])
d = np.array([10,20])
print(c*d)
```

<h1>matplotlib</h1>
그래프, 데이터 시각화를 가능하게 해주는 라이브러리

```python
import numpy as np
import matplotlib.pyplot as plt

#0 ~ 6까지 += 0.1
x = np.arange(0,6, 0.1)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, label = 'sin')
plt.plot(x,y2, linestyle='--', label = 'cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend() # 범례 표시
plt.show()
```
